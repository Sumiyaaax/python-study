```python
from selenium import webdriver
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
import urllib.request as req
```

```python
url = "https://note.com/sumiiyaaax"
response = req.urlopen(url)
parse_html = BeautifulSoup(response, 'html.parser')
```

```python
parse_html
```

```python
print(parse_html.title)
```

```python
print(parse_html.title.string)
```

```python
print(parse_html.find_all('a'))
```

```python
title_lists = parse_html.find_all('a')
```

```python
title_lists[1:10]
```

```python
title_lists[10].string
```

```python
title_lists[10].attrs['href']
```

```python
title_list = []
url_list = []
for i in title_lists:
    title_list.append(i.string)
    url_list.append(i.attrs['href'])
```

```python
title_list
```

```python
url_list
```

```python
df_title_url = pd.DataFrame({'Title':title_list, 'URL':url_list})
```

```python
df_title_url
```

```python
df_notnull = df_title_url.dropna(how='any')
```

```python
df_notnull
```

```python
df_notnull['Title'].str.contains('2020')
```

```python
df_notnull[df_notnull['Title'].str.contains('2020')]
```

```python
df_contain = df_notnull[df_notnull['Title'].str.contains('2020')]
```
